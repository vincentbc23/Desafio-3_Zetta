```python
import pandas as pd
import glob
import os


path = r'../data_focos_incendio' 
all_files = glob.glob(os.path.join(path, "*.csv"))


combined_df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)

combined_df.to_csv('../data_focos_incendio/combined_focos_incendio.csv', index=False)
```
