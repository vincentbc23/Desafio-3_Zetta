```python
df = df.map(lambda x: {'Bioma': 'Desconhecido' if x['Bioma'] is None else x['Bioma']}, num_proc=8)
```

```python
linhas_nulas = df.filter(lambda x: x['Bioma'] is None, num_proc = 4)
linhas_nulas
```

```python
# df['DataHora'] = pd.to_datetime(df['DataHora'])
# resultado = df[df['DataHora'] == '2016-02-06 11:45:00']
# resultado
df.filter(lambda x: x['Bioma'] == 'Desconhecido', num_proc=4).to_pandas()
```

```python
tem_negativos = len(df.filter(lambda x: x['DiaSemChuva'] is not None and x['DiaSemChuva'] < 0 and x['DiaSemChuva'] == -999, num_proc=4)) > 0
tem_negativos
```

```python
tem_negativos = len(df.filter(lambda x: x['DiaSemChuva'] is not None and x['DiaSemChuva'] < 0 and x['DiaSemChuva'] != -999, num_proc=4)) > 0
tem_negativos
```

Ou seja: em DiaSemChuva possui valores negativos, porém todos esses valores negativos são = -999
