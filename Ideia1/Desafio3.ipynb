344233 dessas linhas tem dia sem chuva negativo, como tambem sei que tem valores dessa coluna nulos, vou transformar essas inconsistencias em nulos para depois tratar

```python
df_para_media = df.select_columns(['Bioma', 'DiaSemChuva']).to_pandas()

df_validos = df_para_media[(df_para_media['DiaSemChuva'] >= 0) & (df_para_media['DiaSemChuva'].notna())]

medias_bioma = df_validos.groupby('Bioma')['DiaSemChuva'].mean().to_dict()

medias_bioma
```

```python
def preenche_inconsistencias(x):
    if (x['DiaSemChuva'] is None or x['DiaSemChuva'] < 0) and x['Bioma'] in medias_bioma:
        x['DiaSemChuva'] = medias_bioma[x['Bioma']]
    return x

# Roda o mapeamento usando 4 núcleos do seu processador
df = df.map(preenche_inconsistencias, num_proc=8)
```

```python
df.filter(lambda x: x['DiaSemChuva'] is not None and x['DiaSemChuva'] < 0, num_proc=8)
```

```python
df.to_pandas().isnull().sum()
```

```python
qtd_nulos_diaSemChuva = len(df.filter(lambda x: x['DiaSemChuva'] is None, num_proc=8))
print(f"Total de valores nulos restantes na dias sem chuva: {qtd_nulos_diaSemChuva}")

df_conferencia = df.select_columns(['Bioma', 'DiaSemChuva']).to_pandas()
novas_medias = df_conferencia.groupby('Bioma')['DiaSemChuva'].mean().to_dict()
novas_medias
```

```python
media_caatinga = novas_medias['Caatinga']

linhas_modificadas = df.filter(
    lambda x: x['Bioma'] == 'Caatinga' and x['DiaSemChuva'] == media_caatinga, 
    num_proc=8
)

linhas_modificadas.to_pandas().head()
```

Dias sem chuva tratado com sucesso a partir da media de dias sem chuva por bioma

```python
def preenche_nulos_precipitacao(x):
    if x['Precipitacao'] is None and x['Bioma'] in medias_precipitacao:
        x['Precipitacao'] = medias_precipitacao[x['Bioma']]
    return x

df = df.map(preenche_nulos_precipitacao, num_proc=8)
```

Com o teste feito com a caatinga comprovamos que a mudanca foi feita corretamente

```python
df_para_media_rf = df.select_columns(['Bioma', 'RiscoFogo']).to_pandas()

df_validos_rf = df_para_media_rf[(df_para_media_rf['RiscoFogo'] >= 0) & (df_para_media_rf['RiscoFogo'].notna())]

medias_riscofogo = df_validos_rf.groupby('Bioma')['RiscoFogo'].mean().to_dict()

medias_riscofogo
```

```python
def preenche_riscofogo(x):
    if (x['RiscoFogo'] is None or x['RiscoFogo'] < 0) and x['Bioma'] in medias_riscofogo:
        x['RiscoFogo'] = medias_riscofogo[x['Bioma']]
    return x

df = df.map(preenche_riscofogo, num_proc=8)
```

```python
qtd_nulos_riscoFogo = len(df.filter(lambda x: x['RiscoFogo'] is None, num_proc=8))
print(f"Total de valores nulos restantes na RiscoFogo: {qtd_nulos_riscoFogo}")

df_conferencia = df.select_columns(['Bioma', 'RiscoFogo']).to_pandas()
novas_medias = df_conferencia.groupby('Bioma')['RiscoFogo'].mean().to_dict()
novas_medias
```

RiscoFogo arrumado com sucesso.

```python
df_refinamento = df.select_columns(['Bioma', 'FRP', 'DataHora']).to_pandas()

df_validos = df_refinamento[(df_refinamento['FRP'] >= 0) & (df_refinamento['FRP'].notna())].copy()

plt.figure(figsize=(14, 7))
sns.boxplot(
    data=df_validos, 
    x='Bioma', 
    y='FRP', 
    palette='Set2',
    flierprops={"marker": "o", "markersize": 3, "alpha": 0.5}
)

plt.title('Distribuição e Outliers do FRP por Bioma', fontsize=16)
plt.xlabel('Bioma', fontsize=14)
plt.ylabel('FRP', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Como foi possivel observar o FRP por possuir muitos valores nulos tem muitos outliers, sera usado a mediana por bioma e mes em vez da media por conta disso.

```python
df_refinamento_frp = df.select_columns(['Bioma', 'FRP']).to_pandas()

df_validos_frp = df_refinamento_frp[(df_refinamento_frp['FRP'] >= 0) & (df_refinamento_frp['FRP'].notna())]

medianas_por_bioma = df_validos_frp.groupby('Bioma')['FRP'].median().to_dict()

print("Medianas globais de FRP por Bioma:")
medianas_por_bioma
```

```python
def preenche_frp_mediana_bioma(x):
    if (x['FRP'] is None or x['FRP'] < 0) and x['Bioma'] in medianas_por_bioma:
        x['FRP'] = medianas_por_bioma[x['Bioma']]
    return x

df = df.map(preenche_frp_mediana_bioma, num_proc=8)
```

```python
qtd_nulos_frp = len(df.filter(lambda x: x['FRP'] is None, num_proc=8))
qtd_negativos_frp = len(df.filter(lambda x: x['FRP'] is not None and x['FRP'] < 0, num_proc=8))

print(f"Total de nulos restantes no FRP: {qtd_nulos_frp}")
print(f"Total de negativos (erros) restantes no FRP: {qtd_negativos_frp}")

print("-" * 50)

df_conferencia_frp = df.select_columns(['Bioma', 'FRP']).to_pandas()

resumo_frp = df_conferencia_frp.groupby('Bioma')['FRP'].describe()[['mean', '50%']]

resumo_frp
```

```python
mediana_caatinga = medianas_por_bioma['Caatinga']

linhas_modificadas_frp = df.filter(
    lambda x: x['Bioma'] == 'Caatinga' and x['FRP'] == mediana_caatinga, 
    num_proc=8
)

linhas_modificadas_frp.to_pandas().head()
```

Perfeito, mudancas feitas com sucesso
